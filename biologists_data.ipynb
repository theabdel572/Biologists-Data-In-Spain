```python
import pandas as pd
data = pd.read_csv('53344.csv', sep='\t', header=5, skipfooter=5, names=['Comunidades', 'Total', 'TotalH', 'TotalM', 'Colegiados no jubilados', 'Colegiados no jubiladosH', 'Colegiados no jubiladosM', 'Colegiados jubilados', 'Colegiados jubiladosH', 'Colegiados jubiladosM','No consta', 'No constaH', 'No constaM'], index_col=0, engine='python')
data
```
